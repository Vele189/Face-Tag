```python
import cv2
import face_recognition
import sqlite3
import numpy as np
import os
from IPython.display import clear_output, Image, display
import PIL.Image
import io
```

```python
def register_user(name, age, image_path):
       # Load the image and generate face encoding
       image = face_recognition.load_image_file(image_path)
       encoding = face_recognition.face_encodings(image)[0]

       # Save user details and encoding to the database
       conn = sqlite3.connect('face_recognition.db')
       cursor = conn.cursor()
       cursor.execute("INSERT INTO users (name, age, encoding) VALUES (?, ?, ?)", (name, age, encoding.tobytes()))
       conn.commit()
       conn.close()

   # Example usage

user1 = register_user("Ndamulelo Vele", 20, "images/ndamu.jpg")
user2 = register_user("Donald Trump", 78, "images/trump.jpeg")
user3 = register_user("Aubrey Graham", 38, "images/drake.jpeg")
user4 = register_user("Elon Musk", 53, "images/elon.jpeg")
user5 = register_user("Rihanna Badgirl", 36, "images/rihanna.jpeg")
```

```python
def view_database_contents():
    conn = sqlite3.connect('face_recognition.db')
    cursor = conn.cursor()
    
    cursor.execute("SELECT name, age, encoding FROM users")
    users = cursor.fetchall()
    
    print("Database Contents:")
    print("-" * 50)
    
    for user in users:
        name, age, encoding = user
        encoding_array = np.frombuffer(encoding, dtype=np.float64)
        print(f"Name: {name}")
        print(f"Age: {age}")
        print(f"Face Encoding: {encoding_array}")
        print("-" * 50)
    
    conn.close()
    return users
```

```python
import sqlite3

def remove_user(name):
    # Connect to the SQLite database
    conn = sqlite3.connect('face_recognition.db')
    cursor = conn.cursor()

    # Delete the user from the database
    cursor.execute("DELETE FROM users WHERE name = ?", (name,))
    
    # Commit the changes
    conn.commit()

    # Check if any row was deleted
    if cursor.rowcount > 0:
        print(f"User '{name}' has been removed from the database.")
    else:
        print(f"No user found with the name '{name}'.")

    # Close the connection
    conn.close()

# Example usage
if __name__ == "__main__":
    user_to_remove = input("Enter the name of the user to remove: ")
    remove_user(user_to_remove)
```

```python
def identify_user(image_path):
    # Load the image and generate face encoding
    image = face_recognition.load_image_file(image_path)
    encoding = face_recognition.face_encodings(image)[0]

    # Connect to the database
    conn = sqlite3.connect('face_recognition.db')
    cursor = conn.cursor()

    # Fetch all user encodings from the database
    cursor.execute("SELECT name, age, encoding FROM users")  # Fetch age as well
    users = cursor.fetchall()

    for user in users:
        name = user[0]
        age = user[1]  # Get the age
        db_encoding = np.frombuffer(user[2], dtype=np.float64)  # Adjust index for encoding

        # Compare the encodings
        results = face_recognition.compare_faces([db_encoding], encoding)
        if results[0]:
            print(f"User identified: {name}, Age: {age}")
            return name, age, db_encoding
            

    print("No match found.")
    conn.close()
    return None, None, None 
```

```python
def identify_user_jupyter():
    try:
        # Connect to the database
        conn = sqlite3.connect('face_recognition.db')
        cursor = conn.cursor()
        
        # Fetch all user encodings from the database
        cursor.execute("SELECT name, age, encoding FROM users")
        users = cursor.fetchall()
        
        if not users:
            print("No users found in database. Please add user data first.")
            return
            
        print(f"Loaded {len(users)} users from database")
        
        # Convert encodings from database to numpy arrays
        user_encodings = []
        user_names = []
        user_ages = []
        
        for user in users:
            name = user[0]
            age = user[1]
            db_encoding = np.frombuffer(user[2], dtype=np.float64)
            user_encodings.append(db_encoding)
            user_names.append(name)
            user_ages.append(age)
            
        # Start video capture
        video_capture = cv2.VideoCapture(1)
        if not video_capture.isOpened():
            print("Error: Could not open video capture device")
            return
            
        print("Video capture started. Press 'q' in the output window to quit.")
        
        while True:
            # Capture a single frame
            ret, frame = video_capture.read()
            if not ret:
                print("Failed to capture image")
                break
                
            # Convert the image from BGR to RGB for face_recognition
            rgb_frame = frame[:, :, ::-1]
            
            try:
                # Find all face locations and encodings in the current frame
                face_locations = face_recognition.face_locations(rgb_frame)
                
                if face_locations:
                    face_encodings = face_recognition.face_encodings(rgb_frame, face_locations)
                    
                    # Loop through each face found in the frame
                    for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
                        # Compare the face encoding with the known encodings
                        matches = face_recognition.compare_faces(user_encodings, face_encoding, tolerance=0.6)
                        name = "Unknown"
                        age = "N/A"
                        
                        # If a match is found, get the name and age
                        if True in matches:
                            first_match_index = matches.index(True)
                            name = user_names[first_match_index]
                            age = user_ages[first_match_index]
                            print(f"Match found: {name}, Age: {age}")
                            
                        # Draw a rectangle around the face
                        cv2.rectangle(frame, (left, top), (right, bottom), (0, 255, 0), 2)
                        
                        # Display the name and age below the face
                        text = f"{name}, Age: {age}"
                        cv2.putText(frame, text, (left, bottom + 20), 
                                  cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
                
                # Convert frame to PIL Image for Jupyter display
                frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                pil_img = PIL.Image.fromarray(frame_rgb)
                
                # Convert PIL image to bytes
                buf = io.BytesIO()
                pil_img.save(buf, format='JPEG')
                buf.seek(0)
                
                # Clear previous output and display new frame
                clear_output(wait=True)
                display(Image(data=buf.getvalue()))
                
            except Exception as e:
                print(f"Error processing frame: {e}")
                continue
                
            # Break the loop if 'q' is pressed
            if cv2.waitKey(1) & 0xFF == ord('q'):
                print("Quitting...")
                break
                
    except Exception as e:
        print(f"Error: {e}")
    finally:
        # Release the video capture
        if 'video_capture' in locals():
            video_capture.release()
            cv2.destroyAllWindows()
        conn.close()
```
